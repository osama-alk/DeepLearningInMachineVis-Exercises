# What are fibonachi numbers?

Fibonacci numbers are a sequence of numbers where each number is the sum of the two preceding ones, starting with 0 and 1. The sequence begins: 0, 1, 1, 2, 3, 5, 8, 13, 21, and so on.

A Fibonachi sequene of length n can be easily constructed using the following python code:

First we define the length of the sequence we want to contstuct

```python
n = 10
```

Then we have to start with values 0 and 1 in our list

```python
fib = [0,1]
```

Then we consruct the rest of the elements based on the first 2 elements

```python
n = n - 2 # We alrady have 2 elements in the list
while n > 0 :
    fib.append(fib[-1] + fib[-2])
    n = n - 1
```

Now that we have a list containing the Fibonachi sequence of length n, we can just print it

```python
print(fib)
```

So that is all for how to construct a a Fibonachi sequence of length n
